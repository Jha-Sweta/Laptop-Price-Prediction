```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/drive/MyDrive/Adverk AI 2025  03 Batch/datasets/laptop_data_cleaned.csv")
df
```

```python
df.info()
```

```python
X=df.drop(columns=['Price'])
y=df['Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.15,random_state=14)
```

```python
X.shape,X_train.shape,X_test.shape
```

```python
1040/1224
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import root_mean_squared_error, r2_score, mean_absolute_error
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
```

```python
#step1: Encoding
#step2: LinearRegression
#step1+step2: Pipeline
```

```python
X_train.sample()
```

```python
step1=ColumnTransformer(transformers=[('col_tnf',OneHotEncoder(drop='first'),[0,1,2,4,5])],remainder='passthrough')
step2=LinearRegression()
pipe=Pipeline([('step1',step1),('step2',step2)])
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
print("R2 Score :   ",r2_score(y_test,y_pred))
print("RMSE     :   ",root_mean_squared_error(y_test,y_pred))
print("MAE      :   ",mean_absolute_error(y_pred,y_test))
```

```python
# Lasso
step1=ColumnTransformer(transformers=[('col_tnf',OneHotEncoder(drop='first'),[0,1,2,4,5])],remainder='passthrough')
step2=Lasso(alpha=100)
pipe=Pipeline([('step1',step1),('step2',step2)])
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
print("R2 Score :   ",r2_score(y_test,y_pred))
print("RMSE     :   ",root_mean_squared_error(y_test,y_pred))
print("MAE      :   ",mean_absolute_error(y_pred,y_test))
```

```python
# Ridge
step1=ColumnTransformer(transformers=[('col_tnf',OneHotEncoder(drop='first'),[0,1,2,4,5])],remainder='passthrough')
step2=Ridge(alpha=10)
pipe=Pipeline([('step1',step1),('step2',step2)])
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
print("R2 Score :   ",r2_score(y_test,y_pred))
print("RMSE     :   ",root_mean_squared_error(y_test,y_pred))
print("MAE      :   ",mean_absolute_error(y_pred,y_test))
```

```python
# KNeighborsRegressor
step1=ColumnTransformer(transformers=[('col_tnf',OneHotEncoder(drop='first'),[0,1,2,4,5])],remainder='passthrough')
step2=KNeighborsRegressor()
pipe=Pipeline([('step1',step1),('step2',step2)])
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
print("R2 Score :   ",r2_score(y_test,y_pred))
print("RMSE     :   ",root_mean_squared_error(y_test,y_pred))
print("MAE      :   ",mean_absolute_error(y_pred,y_test))
```

```python
# AdaBoostingRegressor
step1=ColumnTransformer(transformers=[('col_tnf',OneHotEncoder(drop='first'),[0,1,2,4,5])],remainder='passthrough')
step2=AdaBoostRegressor()
pipe=Pipeline([('step1',step1),('step2',step2)])
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
print("R2 Score :   ",r2_score(y_test,y_pred))
print("RMSE     :   ",root_mean_squared_error(y_test,y_pred))
print("MAE      :   ",mean_absolute_error(y_pred,y_test))
```

```python
%%writefile app.py
import streamlit as st
import numpy as np
import pickle

reg_model=pickle.load(open('pipe.pkl','rb'))
df=pickle.load(open('df.pkl','rb'))

st.title("Laptop Price Predictor")

company=st.selectbox("Manufacturer",df['Company'].unique(),index=4)
typename=st.radio("Type of the laptop",df['TypeName'].unique(),index=1,horizontal=True)
cpu=st.selectbox("Processor",df['Cpu'].unique(),index=0)
cpu_speed=st.slider("Processor Clock Speed",min_value=0.8,max_value=3.7,step=0.1,value=2.5)
ram=st.radio("RAM (in GB)",[4,6,8,12,16,24,32,64,128],index=2,horizontal=True)
gpu=st.radio("Graphics Card",df['Gpu'].unique(),index=1,horizontal=True)
os=st.radio("Operating System",df['OpSys'].unique(),index=2,horizontal=True)
weight=st.slider("Weight of the laptop(in kg)",min_value=0.5,max_value=5.0,step=0.5,value=2.0)
ips=st.radio("Does the laptop have IPS Display?",['Yes','No'],index=1,horizontal=True)
touchscreen=st.radio("Does the laptop have a touchscreen?",['Yes','No'],index=1,horizontal=True)
ssd=st.radio("Amount of SSD storage(in GB).(If no SDD and having HDD, then select 0)",[0,128,256,512,1024],index=3,horizontal=True)
hdd=st.radio("Amount of HDD storage(in GB).(If no HDD and having SSD, then select 0)",[0,256,512,1024,2048],index=0,horizontal=True)
ppi=st.slider("Pixel Density",min_value=91,max_value=352,step=1,value=141)

if st.button("PREDICT PRICE"):
    if ips=="Yes":
        ips=1
    else:
        ips=0
    if touchscreen=="Yes":
        touchscreen=1
    else:
        touchscreen=0
    query=np.array([[company,typename,cpu,ram,gpu,os,weight,ips,touchscreen,cpu_speed,hdd,ssd,ppi]])
    op=reg_model.predict(query)
    st.subheader("The estimated price of the laptop with the above mentioned specifications is â‚¹"+str(int(round(op[0],-2))))
```
